```python
#more general form
def Epsi_act(thsx,dt,soilp,D_part,vG):
    psi=vG.psi_thst(thsx,soilp.alpha,soilp.n)
    PSI=psi*9806.38 #m -> Pa
    A=vG.psi_thst(thsx-0.001,soilp.alpha,soilp.n)-vG.psi_thst(thsx+0.001,soilp.alpha,soilp.n)
    B=vG.theta_thst(thsx-0.001,soilp.ts,soilp.tr)-vG.theta_thst(thsx+0.001,soilp.ts,soilp.tr)
    V_part=np.pi/6.*D_part**3
    E_psix=PSI*V_part*(A/B)
    #darcy flux
    k=vG.ku_psi(psi, soilp.ks, soilp.alpha, soilp.n)
    Q=k*psi/D_part
    #time to infiltration of one particle
    t_pinf=D_part/Q
    return np.abs(E_psix)/t_pinf

A=np.arange(0.01,0.99,0.001)
i=0
z1=np.zeros(len(A))
z2=np.zeros(len(A))
z3=np.zeros(len(A))
for x in A: 
    z1[i]=Epsi_act(x,1.,mc.soilmatrix.loc[6],mc.particleD,vG)
    z2[i]=Epsi_act(x,1.,carsel.loc[11],mc.particleD,vG)
    z3[i]=Epsi_act(x,1.,carsel.loc[1],mc.particleD,vG)
    i+=1
```

```python
#now that we have this > get modified advective step length due to friction and psi-dissipation
def s_exchange(thsx,ux,dt,soilp,D_part,m_part,vG,fex=True):
    psi=vG.psi_thst(thsx,soilp.alpha,soilp.n)
    PSI=psi*9806.38 #m -> Pa
    s_proj=ux*dt
    
    ##EXCHANGE##
    A=vG.psi_thst(thsx-0.001,soilp.alpha,soilp.n)-vG.psi_thst(thsx+0.001,soilp.alpha,soilp.n)
    B=vG.theta_thst(thsx-0.001,soilp.ts,soilp.tr)-vG.theta_thst(thsx+0.001,soilp.ts,soilp.tr)
    V_part=np.pi/6.*D_part**3
    E_psix=PSI*V_part*(A/B)
    #darcy flux
    k=vG.ku_psi(psi, soilp.ks, soilp.alpha, soilp.n)
    Q=k*-psi/D_part

    if Q>0.:
        #time to infiltration of one particle
        t_pinf=D_part/Q
        #infiltration depth
        s_inf=Q*dt
        #E of psi dissipation
        E_psi=np.abs(E_psix)*np.amin([dt/t_pinf,1.])
        #F_exchange
        F_exchange1=E_psi/s_inf
        F_exchange=np.abs(E_psix)/D_part
    else:
        F_exchange=0.
        F_exchange1=0.
    
    ##TRANSLATORY ENERGY##
    E_tkin=m_part/2.*ux**2
    F_fric=E_tkin/s_proj
    
    ##Retarded step##
    if fex:
        s_real=E_tkin/(F_fric+F_exchange)
    else:
        s_real=E_tkin/(F_fric+F_exchange1)
            
    
    return s_real


A=np.arange(0.01,0.99,0.001)
ux1=0.203 #Shipitalo and Butt 
ux2=0.0203 #
ux3=0.00608

i=0
y1=np.zeros(len(A))
y2=np.zeros(len(A))
y3=np.zeros(len(A))

y4=np.zeros(len(A))
y5=np.zeros(len(A))
y6=np.zeros(len(A))

y7=np.zeros(len(A))
y8=np.zeros(len(A))
y9=np.zeros(len(A))

for x in A: 
    y1[i]=s_exchange(x,ux1,1.,mc.soilmatrix.loc[6],mc.particleD,mc.particlemass*1000.,vG,False)
    y2[i]=s_exchange(x,ux2,1.,mc.soilmatrix.loc[6],mc.particleD,mc.particlemass*1000.,vG,False)
    y3[i]=s_exchange(x,ux3,1.,mc.soilmatrix.loc[6],mc.particleD,mc.particlemass*1000.,vG,False)
    y4[i]=s_exchange(x,ux1,1.,carsel.loc[11],mc.particleD,mc.particlemass*1000.,vG,False)
    y5[i]=s_exchange(x,ux2,1.,carsel.loc[11],mc.particleD,mc.particlemass*1000.,vG,False)
    y6[i]=s_exchange(x,ux3,1.,carsel.loc[11],mc.particleD,mc.particlemass*1000.,vG,False)
    y7[i]=s_exchange(x,ux1,1.,carsel.loc[1],mc.particleD,mc.particlemass*1000.,vG,False)
    y8[i]=s_exchange(x,ux2,1.,carsel.loc[1],mc.particleD,mc.particlemass*1000.,vG,False)
    y9[i]=s_exchange(x,ux3,1.,carsel.loc[1],mc.particleD,mc.particlemass*1000.,vG,False)
    i+=1

ship1, = plot(A,y1,'-',color=tableau20[0],label='s_real [Shipitalo]')
weil, = plot(A,y2,':',color=tableau20[1],label='s_real [Weiler]')
zeh, = plot(A,y3,ls='dashed',color=tableau20[0],label='s_real [Zehe]')
ship2, = plot(A,y4,'-',color=tableau20[2])
plot(A,y5,':',color=tableau20[3])
plot(A,y6,ls='dashed',color=tableau20[2])
ship3, = plot(A,y7,'-',color=tableau20[4])
plot(A,y8,'-',color=tableau20[5])
plot(A,y9,ls='dashed',color=tableau20[4])

legend([ship1,weil,zeh,ship1,ship2,ship3], ['Shipitalo','Weiler','Zehe','Attert Loam','Carsel SiltLoam','Carsel Clay'],ncol=2,loc=4,fontsize=12,frameon=False)
xlabel('Theta*',fontsize=12)
ylabel('step [m]',fontsize=12)
yscale('log')

title('Advective step reduction due to macropore-matrix\nexchange at different moisture states\n',fontsize=14)
```
