### Discharge of main Attert Stations

```python
Area=pd.DataFrame(np.array([19033276.859635,451211.742,1063268.993,1562816.821,214976.928,92475.551,459105.388,5462202.539,5254594.678,246722920.542,2700000.,4500000.]),
                  index=['Colpach','Weierbach','F','J','L','R','U','X','Z','Useldange','Huewelerbach','Wollefsbach']).T
```

```python
Qfiles=!ls Q-*
```

```python
#get list of all covered catchments
dummyc=[]
for i in np.arange(len(Qfiles)):
    dummyc=np.append(dummyc,Qfiles[i].split('-')[1])
dummycu=np.unique(dummyc)
```

```python
dummycu
```

```python
def read_Q(filename):
    catchment=filename.split('-')[1]
    dummy=pd.read_csv(filename,sep='\t',na_values='---',header=-1)
    dummy.index=pd.to_datetime(dummy[0]+' '+dummy[1],format='%d/%m/%Y %H:%M:%S')
    dummy.columns=['date','time',catchment]
    return dummy.iloc[:,2:]
```

```python
#COLPACH
idx=np.where(dummyc==dummycu[0])[0]
Q_Colpach=read_Q(Qfiles[idx[0]])
for i in np.arange(len(idx))[1:]:
    Q_Colpach=pd.concat([Q_Colpach,read_Q(Qfiles[idx[i]])])
        
```

```python
#HUEWELERBACH
idx=np.where(dummyc==dummycu[1])[0]
Q_Huewelerbach=read_Q(Qfiles[idx[0]])
for i in np.arange(len(idx))[1:]:
    Q_Huewelerbach=pd.concat([Q_Huewelerbach,read_Q(Qfiles[idx[i]])])
        
```

```python
#WOLLEFSBACH
idx=np.where(dummyc==dummycu[8])[0]
Q_Wollefsbach=read_Q(Qfiles[idx[0]])
for i in np.arange(len(idx))[1:]:
    Q_Wollefsbach=pd.concat([Q_Wollefsbach,read_Q(Qfiles[idx[i]])])
        
```

```python
#WEIERBACH
idx=np.where(dummyc==dummycu[7])[0]
Q_Weierbach=read_Q(Qfiles[idx[0]])
for i in np.arange(len(idx))[1:]:
    Q_Weierbach=pd.concat([Q_Weierbach,read_Q(Qfiles[idx[i]])])
        
```

```python
#NIEDERPALLEN
idx=np.where(dummyc==dummycu[2])[0]
Q_Niederpallen=read_Q(Qfiles[idx[0]])
for i in np.arange(len(idx))[1:]:
    Q_Niederpallen=pd.concat([Q_Niederpallen,read_Q(Qfiles[idx[i]])])
```

```python
#PLATEN
idx=np.where(dummyc==dummycu[3])[0]
Q_Platen=read_Q(Qfiles[idx[0]])
for i in np.arange(len(idx))[1:]:
    Q_Platen=pd.concat([Q_Platen,read_Q(Qfiles[idx[i]])])
```

```python
#REICHLANGE
idx=np.where(dummyc==dummycu[4])[0]
Q_Reichlange=read_Q(Qfiles[idx[0]])
for i in np.arange(len(idx))[1:]:
    Q_Reichlange=pd.concat([Q_Reichlange,read_Q(Qfiles[idx[i]])])
```

```python
#SCHWEBICH
idx=np.where(dummyc==dummycu[5])[0]
Q_Schwebich=read_Q(Qfiles[idx[0]])
for i in np.arange(len(idx))[1:]:
    Q_Schwebich=pd.concat([Q_Schwebich,read_Q(Qfiles[idx[i]])])
```

```python
#USELDANGE
idx=np.where(dummyc==dummycu[6])[0]
Q_Useldange=read_Q(Qfiles[idx[0]])
for i in np.arange(len(idx))[1:]:
    Q_Useldange=pd.concat([Q_Useldange,read_Q(Qfiles[idx[i]])])
```

```python
TSx=Attert[['SUM_NN050U','Useldange']].copy()
TSx=TSx.resample('H').iloc[(-isnan(TSx.resample('H').iloc[:,0]) & -isnan(TSx.resample('H').iloc[:,1])).values]
TSx.columns=['precip','Q_usel']
```

```python
q_norm=np.cumsum(TSx.Q_usel/TSx.precip.sum())
p_cum=TSx.precip.cumsum()/TSx.precip.sum()
figure(figsize(7,5))
plot(p_cum.values,q_norm.values)
```

```python
years=np.unique(TSx.index.year)
with sns.axes_style("white"):
    for i in np.arange(len(years)-1):
        TSx_sub=TSx.iloc[(TSx.index>=pd.to_datetime(str(years[i])+'-11-01 00:00:00')) & (TSx.index<=pd.to_datetime(str(years[i+1])+'-10-31 23:59:59')),:]
        if len(TSx_sub)>360*24:
            q_norm=np.cumsum(TSx_sub.Q_usel/TSx_sub.precip.sum())
            p_cum=TSx_sub.precip.cumsum()/TSx_sub.precip.sum()
            plot(p_cum.values,q_norm.values,label=str(years[i+1]))
    plot([0.01,0.6],[0.01,0.6],'--',linewidth=0.5)
    
    spring=p_cum[pd.to_datetime(str(years[i])+'-02-01 00:00:00')]
    summer=p_cum[pd.to_datetime(str(years[i])+'-05-01 00:00:00')]
    autumn=p_cum[pd.to_datetime(str(years[i])+'-08-01 00:00:00')]
    plot([spring,spring],[0.01,0.6],'-',linewidth=0.5,color='grey')
    text(spring+0.01,0.03,'Feb 01',rotation='vertical',color='grey',horizontalalignment='left',verticalalignment='bottom')
    plot([summer,summer],[0.01,0.6],'-',linewidth=0.5,color='grey')
    text(summer+0.01,0.03,'May 01',rotation='vertical',color='grey',horizontalalignment='left',verticalalignment='bottom')
    plot([autumn,autumn],[0.01,0.6],'-',linewidth=0.5,color='grey')
    text(autumn+0.01,0.03,'Aug 01',rotation='vertical',color='grey',horizontalalignment='left',verticalalignment='bottom')
    legend(loc=4)
    axis('equal')
    sns.despine(offset=5, trim=False)
    sns.axlabel('sum P / sum(tot) P','sum Q / sum(tot) P')
    
```

```python
def psum_qsum_plot(TSx,xtitle='Catchment',timeticks=[],interannual=False,lloc=4,figsi=(20,20)):
    if interannual:
        years=np.unique(TSx.index.year)
        if len(TSx.columns)>2:
            fig=plt.figure(figsize=figsi)
            gs = gridspec.GridSpec(int(np.ceil(sqrt(len(years)))),int(np.ceil(sqrt(len(years)))))
        
        for i in np.arange(len(years)-1):
            TSx_sub=TSx.iloc[(TSx.index>=pd.to_datetime(str(years[i])+'-11-01 00:00:00')) & (TSx.index<=pd.to_datetime(str(years[i+1])+'-10-31 23:59:59')),:]
            if (len(TSx_sub)>360*24) & (len(TSx_sub.columns)==2):
                q_norm=np.cumsum(TSx_sub.Q/TSx_sub.P.sum())
                p_cum=TSx_sub.P.cumsum()/TSx_sub.P.sum()
                plot(p_cum.values,q_norm.values,label=str(years[i+1]))
            elif (len(TSx_sub)<=360*24) & (len(TSx_sub.columns)==2):
                plot([0.05,0.05],[0.05,0.05],'.',linewidth=0,label=str(years[i+1]))
                
            elif (len(TSx_sub)>360*24) & (len(TSx_sub.columns)>2):
                subplot(gs[i])
                p_check=0
                for qsub in np.arange(len(TSx_sub.columns))[1:]:
                    TSx_sub2=TSx_sub.resample('H').iloc[(-isnan(TSx_sub.resample('H').iloc[:,0]) & -isnan(TSx_sub.resample('H').iloc[:,qsub])).values,[0,qsub]]
                    if len(TSx_sub2)>350*24:
                        p_check=1
                        q_norm=np.cumsum(TSx_sub2.iloc[:,1]/TSx_sub2.P.sum())
                        p_cum=TSx_sub2.P.cumsum()/TSx_sub2.P.sum()
                        plot(p_cum.values,q_norm.values,label=TSx_sub2.columns[1])
                        
                    else:
                        plot([0.05,0.05],[0.05,0.05],'.',linewidth=0,label=TSx_sub2.columns[1])
                plot([0.05,0.6],[0.05,0.6],'--',linewidth=0.5)
                
                if p_check==1:
                    spring=p_cum[pd.to_datetime(str(years[i+1])+'-02-01 00:00:00')]
                    summer=p_cum[pd.to_datetime(str(years[i+1])+'-05-01 00:00:00')]
                    autumn=p_cum[pd.to_datetime(str(years[i+1])+'-08-01 00:00:00')]
                    plot([spring,spring],[0.01,0.6],'-',linewidth=0.5,color='grey')
                    text(spring+0.01,0.03,'Feb 01',rotation='vertical',color='grey',horizontalalignment='left',verticalalignment='bottom')
                    plot([summer,summer],[0.01,0.6],'-',linewidth=0.5,color='grey')
                    text(summer+0.01,0.03,'May 01',rotation='vertical',color='grey',horizontalalignment='left',verticalalignment='bottom')
                    plot([autumn,autumn],[0.01,0.6],'-',linewidth=0.5,color='grey')
                    text(autumn+0.01,0.03,'Aug 01',rotation='vertical',color='grey',horizontalalignment='left',verticalalignment='bottom')
                            
                if lloc>0:
                    legend(loc=lloc,ncol=2)
                ylim([0.05,0.6])
                axis('equal')
                sns.despine(offset=5, trim=False)
                sns.axlabel('sum P / sum(tot) P','sum Q / sum(tot) P')
                title(str(years[i+1]))
        
    else:
        if len(TSx.columns)==2:
            q_norm=np.cumsum(TSx.Q/TSx.P.sum())
            p_cum=TSx.P.cumsum()/TSx.P.sum()
            plot(p_cum.values,q_norm.values)
        elif len(TSx.columns)>2:
            for qsub in np.arange(len(TSx.columns))[1:]:
                TSx_sub=TSx.resample('H').iloc[(-isnan(TSx.resample('H').iloc[:,0]) & -isnan(TSx.resample('H').iloc[:,qsub])).values,[0,qsub]]
                if len(TSx_sub>363*24):
                    q_norm=np.cumsum(TSx_sub.iloc[:,1]/TSx.P.sum())
                    p_cum=TSx_sub.P.cumsum()/TSx_sub.P.sum()
                    plot(p_cum.values,q_norm.values,label=TSx_sub.columns[1])
                
    plot([0.01,0.6],[0.01,0.6],'--',linewidth=0.5)
    
    if lloc>0:
        legend(loc=lloc,ncol=2)
    axis('equal')
    sns.despine(offset=5, trim=False)
    sns.axlabel('sum P / sum(tot) P','sum Q / sum(tot) P')
    
```

```python
with sns.axes_style("white"):
    figure(figsize=(20,2.5))
    gs = gridspec.GridSpec(1,5)
    i=0
    for catch in np.array(['Useldange','Huewelerbach','Wollefsbach','Weierbach','Colpach']):
        TSx=Attert[['SUM_NN050U',catch]].copy()
        TSx=TSx.resample('H').iloc[((-isnan(TSx.resample('H').iloc[:,0]) & (-isnan(TSx.resample('H').iloc[:,1])))).values]
        TSx.columns=['P','Q']
        if (catch=='Colpach') | (catch=='Weierbach'):
            TSx.P*=corfac
        TSx.Q[TSx.Q<0]=np.nan
        subplot(gs[i])
        psum_qsum_plot(TSx,interannual=True,lloc=2)
        title(catch)
        i+=1
    #savefig('doublemass_catch_leg.pdf')
```

```python
def ET_SJ(Tmax,Tmin,RHmax,RHmin,Rs,u2,J,lat,Elev, alpha, z0):
    #Taken and simplified from package Evapotranspiration >> Danlu Guo <danlu.guo@adelaide.edu.au>
    #Daily Actual Evapotranspiration after Szilagyi, J. 2007, doi:10.1029/2006GL028708
    #Time Series as daily values zoo objects: T[degC], RH[%], Rs[MJ/m2], u2[km/h], J[julian day], 
    #lat[deg], Elev[m]
    #alpha->penman landuse, z0->reference evap surface level[m]
    #Example: ET_SJ(Tmax,Tmin,RHmax,RHmin,Rs,u2,J,lat=49.8,Elev=440.0,alpha = 0.23, z0 = 0.02)
    #Result in [mm/day]
    lat_rad=lat*np.pi/180.
    alphaPT = 1.31
    sigma=4.903e-09
    Gsc=0.082
    lambdax=2.45
    Ta = (Tmax + Tmin)/2
    
    vs_Tmax = 0.6108 * np.exp(17.27 * Tmax/(Tmax + 237.3))
    vs_Tmin = 0.6108 * np.exp(17.27 * Tmin/(Tmin + 237.3))
    vas = (vs_Tmax + vs_Tmin)/2.
    vabar = (vs_Tmin * RHmax/100. + vs_Tmax * RHmin/100.)/2.
    P = 101.3 * ((293. - 0.0065 * Elev)/293)**5.26
    delta = 4098. * (0.6108 * np.exp((17.27 * Ta)/(Ta + 237.3)))/((Ta + 237.3)**2.)
    gamma = 0.00163 * P/lambdax
    d_r2 = 1. + 0.033 * np.cos(2. * np.pi/365. * J)
    delta2 = 0.409 * np.sin(2. * np.pi/365. * J - 1.39)
    w_s = np.arccos(-np.tan(lat_rad) * np.tan(delta2))
    N = 24./np.pi * w_s
    R_a = (1440./np.pi) * d_r2 * Gsc * (w_s * np.sin(lat_rad) * np.sin(delta2) + np.cos(lat_rad) * np.cos(delta2) * np.sin(w_s))
    R_so = (0.75 + (2. * 10.**-5.) * Elev) * R_a
    
    R_nl = sigma * (0.34 - 0.14 * np.sqrt(vabar)) * ((Tmax + 273.2)**4. + (Tmin + 273.2)**4.)/2. * (1.35 * Rs/R_so - 0.35)
    R_nsg = (1. - alpha) * Rs
    R_ng = R_nsg - R_nl
    
    f_u = 0.26*(1.+0.54*u2) #u2 in m/s
    #f_u = 0.26*(1.+0.86*u2) #u2 in m/s
    
    #f_u = 1.313 + 1.381 * u2 #version 1956
    #f_u = 2.626 + 1.381 * u2 #version 1948
    
    Ea = f_u * (vas - vabar)
    Epenman_Daily = delta/(delta + gamma) * (R_ng/lambdax) + gamma/(delta + gamma) * Ea
    
    T_e = Ta
    for i in np.arange(99999):
        v_e = 0.6108 * np.exp(17.27 * T_e/(T_e + 237.3))
        T_enew = Ta - 1./gamma * (1. - R_ng/(lambdax * Epenman_Daily)) * (v_e - vabar)
        deltaT_e = T_enew - T_e
        if type(deltaT_e)==np.float64:
            maxdeltaT_e = np.abs(deltaT_e)
        else:
            maxdeltaT_e = np.abs(np.max(deltaT_e[-np.isnan(deltaT_e)]))
        T_e = T_enew
        if (maxdeltaT_e < 0.01):
            break
    
    deltaT_e = 4098. * (0.6108 * np.exp((17.27 * T_e)/(T_e + 237.3)))/((T_e + 237.3)**2.)
    E_PT_T_e = alphaPT * (deltaT_e/(deltaT_e + gamma) * R_ng/lambdax)
    E_SJ_Act_Daily = 2. * E_PT_T_e - Epenman_Daily
    E_SJ_Act_Daily[E_SJ_Act_Daily>0.]=0
    
    return(-E_SJ_Act_Daily)
```

```python
#define GOF function

def GOF(o,s):
    import errlib as EL #errorlib as python file in folder
    import scipy.stats as sps
    #Goodness of Fit Measures
    
    idxx=(-isnan(o) & -isnan(s))
    o=o[idxx]
    s=s[idxx]
    [slope, intercept, r_value, p_value, std_err] = sps.stats.linregress(o,s)
    gof_val=np.array([r_value, std_err])
    #R correlation coefficient
    #R2 Standard error of the estimate
    gof_val=np.append(gof_val,np.array(EL.KGE(o,s)))
    #kge: Kling-Gupta Efficiency
    #cc: correlation 
    #alpha: ratio of the standard deviation
    #beta: ratio of the mean
    gof_val=np.append(gof_val,EL.L(o,s))
    #L: Likelihood
    gof_val=np.append(gof_val,EL.NS(o,s))
    #NSE: Nash Sutcliffe efficient coefficient
    gof_val=np.append(gof_val,EL.mae(o,s))
    #mae: Mean Absolute Error
    gof_val=np.append(gof_val,EL.rmse(o,s))
    #rmse: Root Mean Squared Error
    gof_val=np.append(gof_val,np.array(sps.spearmanr(s,o, axis=None)))
    #spear: spearman rank correlation
    #p_spear: p-value
    gof_val=np.append(gof_val,np.array(sps.kendalltau(s,o)))
    #tau: kendalls tau
    #p_kend: p-value
    gof_val=np.append(gof_val,np.array(sps.pearsonr(s,o)))
    #pear: Pearson correlation coefficient
    #p_pear: p-value
    param_names=['cor_1','std_err','KGE','cor','alpha','beta','L','NSE','mae','rmse','spearman','p_spear','kendall','p_kend','pearson','p_pear']
    return pd.DataFrame(data=gof_val,index=param_names,dtype=float64).T
```

### Further development of the ET model

```python
def ET_SJ(data,lat,Elev, alpha, z0,windref=True,windfunction_ver='1948'):
    '''Evapotranspiration after Szilagyi and Josza 2008
       taken from the R-package Evapotranspiration of Danlu Guo <danlu.guo@adelaide.edu.au>
       
       data is expected as pandas dataframe with:
       Tmin (Air Temperature, degCelsius)
       Tmax (Air Temperature, degCelsius)
       RHmin (Relative Humidity, %)
       RHmax (Relative Humidity, %)
       Rs (Global Radiation, MJ/m2) (if one value is nan -> all values are estimated)
       u2 (Wind Velocity, m/s) (if one value is nan -> wind will not be respected)

       lat (latitude of station, degree)
       Elev (elevation above MSL, m)
       alpha (penman landuse, -) 
       z0 (reference evap surface level, m)
       
       windref (False if z0 is not measured at 2m)
       windfunction_ver (Penman version 1948 or 1956)
    '''

    #define constants
    ETconst=pd.DataFrame(np.array([2.45,     4.903e-09,   0.082,   -34.9211,  -0.6094882,  0.23, 0.5, 48,   10, 1.2,  0.001013, 0, 0.14,   1.26,    2.4, 28, 1, 11.9, -0.15, -0.25, -0.0107, 0.81917, -0.0040922, 1.0705, 0.065649, -0.0059684, -0.0005967, 0.66,    0.92,      285.8, 17.27,   237.3,  5.67e-08, 28.5,     14, 1.2]),
                         index=['Lambda', 'sigma',     'Gsc',   'lat',      'lat_rad',    'As',   'bs',  'Elev', 'z',  'Roua', 'Ca',       'G', 'alphaA', 'alphaPT', 'ap',  'fz', 'b0', 'a_0',  'b_0',   'c_0',   'd_0',     'e0',      'e1',         'e2',     'e3',       'e4',         'e5',         'gammaps', 'epsilonMo', 'PA',    'alphaMo', 'betaMo', 'sigmaMo',  'lambdaMo', 'b1', 'b2' ]).T

    #modify constants if given
    if lat!=ETconst.lat.values:
        ETconst.lat=lat
        ETconst.lat_rad=lat*np.pi/180.
    if Elev!=ETconst.Elev.values:
        ETconst.Elev=Elev


    ##calculation
    # average T
    Ta = (data.Tmax + data.Tmin)/2
    # saturation pressure
    vs_Tmax = 0.6108 * exp(17.27 * data.Tmax/(data.Tmax + 237.3))
    vs_Tmin = 0.6108 * exp(17.27 * data.Tmin/(data.Tmin + 237.3))
    vas = (vs_Tmax + vs_Tmin)/2
    # vapour pressure deficit
    vabar = (vs_Tmin * data.RHmax/100 + vs_Tmax * data.RHmin/100)/2
    
    P = 101.3 * ((293 - 0.0065 * ETconst.Elev.values)/293)**5.26
    delta = 4098 * (0.6108 * exp((17.27 * Ta)/(Ta + 237.3)))/((Ta + 237.3)**2)
    gamma = 0.00163 * P/ETconst.Lambda.values
    d_r2 = 1 + 0.033 * np.cos(2 * np.pi/365 * data.index.dayofyear)
    delta2 = 0.409 * np.sin(2 * np.pi/365 * data.index.dayofyear - 1.39)
    w_s = np.arccos(-np.tan(ETconst.lat_rad.values) * np.tan(delta2))
    N = 24/np.pi * w_s
    R_a = (1440/np.pi) * d_r2 * ETconst.Gsc.values * (w_s * np.sin(ETconst.lat_rad.values) * np.sin(delta2) + np.cos(ETconst.lat_rad.values) * np.cos(delta2) * np.sin(w_s))
    R_so = (0.75 + (2 * 10**-5) * ETconst.Elev.values) * R_a
    
    if any(data.Rs == np.nan):
        R_s = (ETconst.As.values + ETconst.bs.values * (data.n/N)) * R_a
    else:
        R_s = data.Rs
    
    if any(data.u2 != np.nan):
        if (windref == False):
            data.u2 = data.u2 * log(2/z0)/log(ETconst.z/z0)
        
        R_nl = ETconst.sigma * (0.34 - 0.14 * np.sqrt(vabar)) * ((data.Tmax + 273.2)**4 + (data.Tmin + 273.2)**4)/2 * (1.35 * R_s/R_so - 0.35)
        R_nsg = (1 - alpha) * R_s
        R_ng = R_nsg - R_nl
    
        if (windfunction_ver == "1948"):
            f_u = 0.26*(1.+0.54*u2) #2.626 + 1.381 * u2
        else:
            f_u = 0.26*(1.+0.86*u2) #1.313 + 1.381 * u2
        
        Ea = f_u * (vas - vabar)
        Epenman_Daily = delta/(delta + gamma) * (R_ng/ETconst.Lambda) + gamma/(delta + gamma) * Ea

    else:
        R_nl = ETconst.sigma * (0.34 - 0.14 * np.sqrt(vabar)) * ((data.Tmax + 273.2)**4 + (data.Tmin + 273.2)**4)/2 * (1.35 * R_s/R_so - 0.35)
        R_nsg = (1 - alpha) * R_s
        R_ng = R_nsg - R_nl
        RHmean = (data.RHmax + data.RHmin)/2
        Epenman_Daily = 0.047 * R_s * np.sqrt(Ta + 9.5) - 2.4 * (R_s/R_a)**2 + 0.09 * (Ta + 20) * (1 - RHmean/100)
    
    #iterate surface temperature
    T_e = Ta
    for i in np.arange(99999):
        v_e = 0.6108 * exp(17.27 * T_e/(T_e + 237.3))
        T_enew = Ta - 1/gamma * (1 - R_ng/(ETconst.Lambda * Epenman_Daily)) * (v_e - vabar)
        deltaT_e = T_enew[-isnan(T_enew)] - T_e
        maxdeltaT_e = np.abs(max(deltaT_e))
        T_e = T_enew
        if (maxdeltaT_e < 0.01):
            break
    
    deltaT_e = 4098. * (0.6108 * exp((17.27 * T_e)/(T_e + 237.3)))/((T_e + 237.3)**2)
    E_PT_T_e = ETconst.alphaPT * (deltaT_e/(deltaT_e + gamma) * R_ng/ETconst.Lambda)
    E_SJ_Act_Daily = 2 * E_PT_T_e - Epenman_Daily
    return pd.DataFrame(E_SJ_Act_Daily,index=data.index)
    
```
