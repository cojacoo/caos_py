```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
import matplotlib.gridspec as gridspec
```

```python
cobra=pd.read_csv('../cobra_cores.csv',sep=',',skiprows=1)
```

```python
cobra['RealDepth']=0.0
```

```python
#stack cores (no depth correction applied here!)
for corenow in cobra.Core.unique():
    cobra.loc[(cobra.Core==corenow)&(cobra.Incr==1),'RealDepth']=cobra.loc[(cobra.Core==corenow)&(cobra.Incr==1)].Depth.values
    for incrnow in np.arange(2)+2:
        cobra.loc[(cobra.Core==corenow)&(cobra.Incr==incrnow),'RealDepth']=(cobra.loc[(cobra.Core==corenow)&(cobra.Incr==incrnow)].Depth+cobra.loc[(cobra.Core==corenow)&(cobra.Incr==incrnow-1)].RealDepth.max()).values
```

```python
#cobra.describe().T
```

```python
cobra.index=-cobra.RealDepth
```

```python
#plot data
with sns.axes_style("white"):
    for c in cobra.Core.unique():
        cobra.loc[cobra.Core==c,['6.3 mm','2 mm','0.63 mm','0.05 mm','<0.05 mm']].iloc[::-1].plot(kind='barh', stacked=True)
        sns.despine(offset=5, trim=False)
```

```python
with sns.axes_style("white"):
    fig=plt.figure(figsize=(15,5))
    gs = gridspec.GridSpec(1, 2)
    
    subplot(gs[0])
    for c in cobra.Core.unique():        
        plot(cobra.loc[cobra.Core==c,'6.3 mm'],-cobra.loc[cobra.Core==c,'RealDepth'],label=c+'_skel')
    sns.despine(offset=5, trim=False)
    plt.legend()
    sns.axlabel('coarse fraction >6.3mm [g/g]','depth [m]')
        
    subplot(gs[1])
    for c in cobra.Core.unique():        
        plot(cobra.loc[cobra.Core==c,'<0.05 mm'],-cobra.loc[cobra.Core==c,'RealDepth'],label=c+'_silt')
    sns.despine(offset=5, trim=False)
    plt.legend()
    sns.axlabel('fine fraction <0.05mm [g/g]','depth [m]')

```
