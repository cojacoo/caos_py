# Analysis of Event 21-06-13 in Holtz Experimental Basin, Attert
We analyse this event for an estimation of catchment response and observed velocities of event reaction.
The general idea is, to use the joint PDF of distances and observed gauge response to define a PDF of velocities.

### Convert the data into PDFs

```python
#get the distribution of distances from stream in U catchment
Udist2stream=pd.read_csv('U_dist2stream.txt',delimiter=' ')
subplot(121)
plot(cumsum(Udist2stream.area)/sum(Udist2stream.area),Udist2stream[['V1','V2']].mean(axis=1))
#plot(Udist2stream.area,Udist2stream[['V1','V2']].mean(axis=1))
xlabel("areal share")
ylabel('distance from stream [m]')
subplot(122)
plot(Udist2stream[['V1','V2']].mean(axis=1),Udist2stream.area/np.sum(Udist2stream.area))
tight_layout()
```

#### consider the discharge as jPDF of the marginals 'distance from stream' and 'flow velocity'.

```python
#binned probability of distances
bins=np.around(Udist2stream['V2'].max(),decimals=-2)/20.
distances=np.arange(20)*bins
#binned probability of timings
bins=len(hevent0.QUmm)/20
discharges=np.arange(20)*bins

cumprob_dist=np.zeros(20)
prob_Q=np.zeros(20)

for i in np.arange(20):
    idx=np.where(Udist2stream[['V1','V2']].mean(axis=1)>=distances[i])[0][0]
    cumprob_dist[i]=sum(Udist2stream.iloc[:idx].area)/Udist2stream.area.sum()
    
    prob_Q[i]=hevent0.QUmm[int(discharges[i])]
    
prob_Q/=sum(prob_Q)
    
```

We know that v=s/t and s=v*t and t=s/v.

We can thus use this to link the marginal of the distances to the joint of the response.

This will then allow the determination of the secon marginal of the velocities.

```python
#construct the probability field of the distances and timings which will be used to derrive the PDF of the velocities
jointprob=np.zeros((19,19))
for i in np.arange(19):
    jointprob[i,:]=diff(cumprob_dist)*prob_Q[i]
    #multiplication of all bin probabilities

#check consistency
sum(jointprob)
```

#Apply joint PDF of timing to apparent velocities
I calculate the velocities of the 2D bins by v=s/t. Then I again bin the resulting velocities and cumulate the probabilities of each of these bins in order to derive the PDF of observed velocities. As such, I derive the PDF of the velocities which is the second independent marginal to the P(distances) resulting in the jPDF(timing).

```python
#get pdf of apparent velocities (distance/timing)
bin_distance=diff(distances)[0] #'m'
bin_timing=diff(discharges)[0]*5*60 #'s'

#get velocities of all bins
distancesx=bin_distance*np.arange(20)[1:]
timings=bin_timing*np.arange(20)[1:]

velocities=np.zeros((19,19))
for i in np.arange(19):
    velocities[i,:]=distancesx/timings[i]

#assign velocities respective probability
#and plot the distribution

plot(ravel(velocities),ravel(jointprob),'.')
xscale('log')
```

```python
#bin velocities and cumulate probablities
#plot(log(ravel(velocities)))
velocitiesxlog=np.arange(-2,-9,-0.25)
velocitiesxprob=np.zeros(len(velocitiesxlog)-1)
for i in np.arange(len(velocitiesxlog))[1:]:
    idx=np.where((np.log(ravel(velocities))>velocitiesxlog[i]) & (np.log(ravel(velocities))<=velocitiesxlog[i-1]))[0]
    velocitiesxprob[i-1]=sum(ravel(jointprob)[idx])

plot(exp(velocitiesxlog[1:]),velocitiesxprob)
xscale('log')
title('PDF of velocities in Holtz Experimental Basin\nEvent 21-06-2013')
xlabel('Velocity [m/s]')
ylabel('Probability')
```

#Note:
This PDF of velocities has a number of assuptions:

* the whole event water is resembled
* only event water is moving through the system
* the velocity of water is constant over time and distance.

Moreover we take the assumption, that it is justified to get the jPDF(timing) by simple multiplication of the 1D PDF(timing) with the 1D PDF(distances). However, this is just one possible filter, we can apply as some kind of uninformative prior.

It is apparent that most of these assuptions do not agree with our observations. However, when keeping in mind, that we are considering the fast and medium part of the velocity spectrum which is likely traveling through structures and may be less effected by retention processes in the soil matrix, this gives reasonable estimates.
