```python
import pandas as pd
import numpy as np
```

### Read data and prepare data frame

```python
#read data
QU=pd.read_csv('../../../caos/DATA/Colpach/S_U/S_U_CTDY_old.dat',delimiter='\t',index_col='TIMESTAMP',na_values='NaN')
QR=pd.read_csv('../../../caos/DATA/Colpach/S_R/S_R_CTD.dat',delimiter='\t',index_col='TIMESTAMP',na_values='NaN')
Prec=pd.read_csv('../../../caos/DATA/Colpach/S_I/S_I_rain.dat',delimiter='\t',index_col='TIMESTAMP',na_values='NaN')
QU = QU.set_index(pd.DatetimeIndex(QU.index))
QR = QR.set_index(pd.DatetimeIndex(QR.index))
Prec = Prec.set_index(pd.DatetimeIndex(Prec.index))
```

```python
#read new Q_U data <-> but this is in 30min! 
QUx=pd.read_csv('../../../caos/DATA/Colpach/S_U/S_U_CTDY.dat',delimiter='\t',index_col='TIMESTAMP',na_values='NaN')
QUx = QUx.set_index(pd.DatetimeIndex(QUx.index))
```

```python
QUx.plot()
```

```python
#merge holtzbach time series
#holtzbach = pd.concat([QR['H2OLevl_X_mm_Avg'],QUx['H2OLevl_Y_mm_Avg'],Prec['rain_mm_Tot.1.']],axis=1,join='outer')
holtzbach = pd.concat([QR['H2OLevl_X_mm_Avg'],QU['H2OLevl_Y_mm_Avg'],Prec['rain_mm_Tot.1.']],axis=1,join='outer')
holtzbach.columns = ['QR','QU','Prec']
holtzbach.plot()
```

```python
#apply rating curves and convert units
h0=16.
holtzbach['QUmm']=(1.4*((holtzbach.QU-h0)/1000.)**(5./2.))*1000.  #[l/s]
holtzbach['QUmm']*=420./459105.4     # *60*5/Area_U [mm/5min]

h0=30.5
holtzbach['QRmm']=-(0.775*((holtzbach.QR-h0)/1000)**(2.47))*1000 
holtzbach['QRmm']*=420./92475.55     # *60*5/Area_R [mm/5min]
```

```python
#binned probability of distances
bins=np.around(Udist2stream['V2'].max(),decimals=-2)/20.
distances=np.arange(20)*bins
#binned probability of timings
bins=(sum(hevent0.QUmm))/20.
discharges=np.arange(20)*bins

cumprob_dist=np.zeros(20)
cumprob_Q=np.zeros(20)

for i in np.arange(20):
    idx=np.where(Udist2stream[['V1','V2']].mean(axis=1)>=distances[i])[0][0]
    cumprob_dist[i]=sum(Udist2stream.iloc[:idx].area)/Udist2stream.area.sum()
    
    idx=np.where(cumsum(hevent0.QUmm)>=discharges[i])[0][0]
    cumprob_Q[i]=(idx-rainmean)/1000.
    
jointprob=np.zeros((19,19))
for i in np.arange(19):
    jointprob[i,:]=diff(cumprob_dist)*diff(cumprob_Q)[i]

imshow(jointprob)
colorbar()
```

```python
import sys
sys.path.append('./event_selection/')
from Event_separation import Event_separation
import run_event_selection
#event separation by Ralf Loritz KIT Hydrology
```

```python
evs = Event_separation()
#Find the preliminary events, last value as integer is the rainfall threshold, last value is the minimal timeperiod for a dry event
df_ev = evs.prec_selection(prec=holtzbach.copy(), catchment_nr='Prec', length=12*24, threshold=2)
```

```python
#Split events in single series
df_ts, df_te = evs.split_events(prec=holtzbach.copy(), catchment_nr='Prec', idx_rain=df_ev)
```

```python
#Find a new startpoint if the ts[n-1] than ts[n]
ts_list = evs.find_new_ts(df_ts=df_ts, runo=holtzbach.copy(),catchment_nr='QU')
        
```

```python
# EXTRA routine to remove bad events
# Removes events if they are only 1 hour long after new ts is found
# or if the runoff[ts] is higher than runoff[te + 1] 
for x in xrange(0, len(df_ts)):
    if (df_te[x] - ts_list[x]) <= 1:
        np.delete(df_te, x)
        np.delete(ts_list, x)
    if holtzbach['QU'][df_te[x].astype(int) + 1] - holtzbach['QU'][ts_list[x].astype(int)] < 0:
        np.delete(df_te, x)
        np.delete(ts_list, x)
#Find a new startpoint if the ts[-1+n] is 20% lower than ts
te_list, reached_k = evs.find_new_te(runo=holtzbach.copy(), catchment_nr='QU',ts_new=ts_list, df_te=df_te, delta_k_threshold=0.00001)
```

```python
# Safety check if ts index is later than te-1 index
# and if a event is shorter than 10 hrs
ts, te = evs.check_duration(ts_list, te_list, threshold=6) 
```
