```python
#load profile file
def loadGPR(filex,res):
    GPR_dummy=np.loadtxt(filex)
    GPR_z=np.ones(res)
    wx=np.amin([shape(GPR_dummy)[1],shape(GPR_z)[1]])
    wz=np.amin([shape(GPR_dummy)[0],shape(GPR_z)[0]])
    GPR_z[:wz,:wx]=GPR_dummy[:wz,:wx]
    return GPR_z

def loadGPRpos(posis,linex):
    GPRpos=np.loadtxt(posis[linex],skiprows=1)
    return GPRpos
```

```python
def P_GPR(i):
    import scipy.ndimage.filters as spf
    GPR_dummy=np.zeros(np.append(res, 8))
    GPR_bg=np.zeros(res)
    for k in np.arange(8):
        GPR_in=loadGPR('./GPR_data/line_data/'+files[k+(i*8)],res)
        GPR_dummy[:,:,k]=spf.median_filter(GPR_in,3)
    
    for k in np.arange(len(GPR_dummy[:,0,0])):
        for j in np.arange(len(GPR_dummy[0,:,0])):
            if len(np.where(diff(GPR_dummy[k,j,:])<=-0.05)[0])==0:
                GPR_bg[k,j]=9
            else:
                if i>0:
                    GPR_bg[k,j]=np.amax([2,np.where(diff(GPR_dummy[k,j,:])<=-0.05)[0][0]+2])
                else:
                    GPR_bg[k,j]=np.where(diff(GPR_dummy[k,j,:])<=-0.05)[0][0]
    
    return GPR_bg

def plt_GPR(kx,i,GPR_bg):
    ''' Plot GPR Profile with discrimination of old/new water
        kx: scene number (2=3rd scene)
        i:  profile number (0=1st profile)
        GPR_bg: mask matrix 
    '''
    import scipy.ndimage.filters as spf
    GPR_dummy=np.zeros(np.append(res, 2))
    for k in np.arange(2):
        GPR_in=loadGPR('./GPR_data/line_data/'+files[kx+k+(i*8)],res)
        GPR_dummy[:,:,k]=spf.median_filter(GPR_in,3)
    
    if kx==0:
        plt.imshow(-GPR_dummy[:,:,0],cmap='Greens',aspect=1,extent=[0,10,-4,0],vmin=-1., vmax=0.)
        sns.despine(offset=5, trim=False)
        plt.yticks([0, -1, -2, -3, -4])
        text(-0.3,-0.4,0,horizontalalignment='right',color='#366AC3')
        text(-0.3,-1.4,int(np.round(1./(v_soil_eff*0.5))),horizontalalignment='right',color='#366AC3')
        text(-0.3,-2.4,int(np.round(2./(v_soil_eff*0.5))),horizontalalignment='right',color='#366AC3')
        text(-0.3,-3.4,int(np.round(3./(v_soil_eff*0.5))),horizontalalignment='right',color='#366AC3')
        text(7.5, 0.1,'time: '+files[kx+k+(i*8)].split()[3].split('h')[0])
        text(0.1, 0.1,'Profile '+str(i+1))
        sns.axlabel('lateral position on profile line [m]','')
        text(-0.72,-1.5,'TWT [ns]',horizontalalignment='right',verticalalignment='bottom',rotation=90,color='#366AC3')
        text(-0.72,-3.5,'est. depth [m]',horizontalalignment='right',verticalalignment='bottom',rotation=90,color='#000000')
        
    else:
        GPR_g = np.ma.array(-GPR_dummy[:,:,0], mask=GPR_bg<=kx)
        GPR_b = np.ma.array(-GPR_dummy[:,:,1], mask=GPR_bg>kx)
        plt.imshow(GPR_g,cmap='Greens',aspect=1,extent=[0,10,-4,0],vmin=-1., vmax=0.,interpolation='none')
        plt.imshow(GPR_b,cmap='Blues',aspect=1,extent=[0,10,-4,0],vmin=-1., vmax=0.,interpolation='none')
        sns.despine(offset=5, trim=False)
        plt.yticks([0, -1, -2, -3, -4])
        text(-0.3,-0.4,0,horizontalalignment='right',color='#366AC3')
        text(-0.3,-1.4,int(np.round(1./(v_soil_eff*0.5))),horizontalalignment='right',color='#366AC3')
        text(-0.3,-2.4,int(np.round(2./(v_soil_eff*0.5))),horizontalalignment='right',color='#366AC3')
        text(-0.3,-3.4,int(np.round(3./(v_soil_eff*0.5))),horizontalalignment='right',color='#366AC3')
        text(7.5, 0.1,'time: '+files[kx+k+(i*8)].split()[3].split('h')[0])
        text(0.1, 0.1,'Profile '+str(i+1))
        sns.axlabel('lateral position on profile line [m]','')
        text(-0.72,-1.5,'TWT [ns]',horizontalalignment='right',verticalalignment='bottom',rotation=90,color='#366AC3')
        text(-0.72,-3.5,'est. depth [m]',horizontalalignment='right',verticalalignment='bottom',rotation=90,color='#000000')
```

### Plot of GPR profile and TDR log (absoulte moisture, not d(theta))

```python
def roundhour(tim):
    AA = pd.to_datetime(tim.year*1000000 + tim.month*10000 + tim.day*100 + tim.hour, format='%Y%m%d%H')
    return [AA, AA+timedelta(1./24.)]

def plt_GPR_trime(kx,i,GPR_bg,TDRtube,stretch,caliparam=(v_soil_eff*0.5*sampling),latadj=-0.3,vertadj=-0.4):
    ''' Plot GPR Profile with discrimination of old/new water
        kx: scene number (2=3rd scene)
        i:  profile number (0=1st profile)
        GPR_bg: mask matrix 
        TDRtube: number of TDR tube
        stretch: reference time like pd.to_datetime('2013-06-21 22:21:00')
        caliparam: calibartion parameter for depth estimate (depth of one pixel [m])
    '''
    import scipy.ndimage.filters as spf
    GPR_dummy=np.zeros(np.append(res, 2))
    tube=['T'+str(TDRtube),'t'+str(TDRtube)]
    
    #prepare GPR data
    for k in np.arange(2):
        GPR_in=loadGPR('./GPR_data/line_data/'+files[kx+k+(i*8)],res)
        GPR_dummy[:,:,k]=spf.median_filter(GPR_in,3)
    
    #prepare trime data
    #find closest stretch
    ref=posi[posi.Point==tube[0]].values[0][[2,1]]
    dist=np.zeros(len(GPR1pos))
    for ix in np.arange(len(GPR1pos)):
        dist[ix]=np.linalg.norm(GPR1pos[ix,:2]-ref)
    idx=np.argmin(dist)*xres
    
    #find closest trime series
    AA=trime[trime.MeasPoint==tube[1]].index
    tdist_min=AA[1]-stretch
    for ix in np.arange(len(AA)):
        tdist=AA[ix]-stretch
        if tdist.seconds<tdist_min.seconds:
            tdist_min=tdist
            tdist_i=ix

    tx=roundhour(AA[tdist_i])
    trime_sub=trime[(trime.index>tx[0]) & (trime.index<tx[1]) & (trime.MeasPoint==tube[1])]
    trime_zmin=-0.01*trime_sub['RealDepth'].max()-0.09
    
    fig=plt.figure(figsize=(15,5))
    gs = gridspec.GridSpec(1, 2, width_ratios=[10,1] ,hspace=0.02)
    
    subplot(gs[0])
    #plot GPR data
    if kx==0:
        plt.imshow(-GPR_dummy[:,:,0],cmap='Greens',aspect=1,extent=[0,10,-4,0],vmin=-1., vmax=0.)
        sns.despine(offset=5, trim=False)
        plt.yticks([0, -1, -2, -3, -4])
        text(0.1, 0.1,'Profile '+str(i+1)+', time: '+files[0+(i*8)].split()[3].split('h')[0])
        text(latadj,-0+vertadj,0,horizontalalignment='right',color='#366AC3')
        text(latadj,-1+vertadj,int(np.round(1./(v_soil_eff*0.5))),horizontalalignment='right',color='#366AC3')
        text(latadj,-2+vertadj,int(np.round(2./(v_soil_eff*0.5))),horizontalalignment='right',color='#366AC3')
        text(latadj,-3+vertadj,int(np.round(3./(v_soil_eff*0.5))),horizontalalignment='right',color='#366AC3')
        sns.axlabel('lateral position on profile line [m]','')
        text(latadj*2.6,-1.5,'TWT [ns]',horizontalalignment='right',verticalalignment='bottom',rotation=90,color='#366AC3')
        text(latadj*2.6,-3.5,'est. depth [m]',horizontalalignment='right',verticalalignment='bottom',rotation=90,color='#000000')
        
        
    else:
        GPR_g = np.ma.array(-GPR_dummy[:,:,0], mask=GPR_bg<=kx)
        GPR_b = np.ma.array(-GPR_dummy[:,:,1], mask=GPR_bg>kx)
        plt.imshow(GPR_g,cmap='Greens',aspect=1,extent=[0,10,-4,0],vmin=-1., vmax=0.,interpolation='none')
        plt.imshow(GPR_b,cmap='Blues',aspect=1,extent=[0,10,-4,0],vmin=-1., vmax=0.,interpolation='none')
        sns.despine(offset=5, trim=False)
        plt.yticks([0, -1, -2, -3, -4])
        #text(0.1, 0.1,'Profile '+str(i+1)+', event: '+files[kx+k+(i*8)].split()[3].split('h')[0]+', '+stretch.ctime())
        text(0.1, 0.1,'GPR Profile '+str(i+1)+', time: '+files[kx+k+(i*8)].split()[3].split('h')[0])
        text(latadj,-0+vertadj,0,horizontalalignment='right',color='#366AC3')
        text(latadj,-1+vertadj,int(np.round(1./(v_soil_eff*0.5))),horizontalalignment='right',color='#366AC3')
        text(latadj,-2+vertadj,int(np.round(2./(v_soil_eff*0.5))),horizontalalignment='right',color='#366AC3')
        text(latadj,-3+vertadj,int(np.round(3./(v_soil_eff*0.5))),horizontalalignment='right',color='#366AC3')
        sns.axlabel('lateral position on profile line [m]','')
        text(latadj*2.6,-1.5,'TWT [ns]',horizontalalignment='right',verticalalignment='bottom',rotation=90,color='#366AC3')
        text(latadj*2.6,-3.5,'est. depth [m]',horizontalalignment='right',verticalalignment='bottom',rotation=90,color='#000000')

        
    #add time location
    plt.plot((idx, idx), (trime_zmin, 0.),  c='#EF5F3A', lw=3) 
    #text(idx-0.2, 0.1,tube[0]+', '+trime_sub.index[0].ctime(),color='#EF5F3A')
    text(idx-0.2, 0.1,tube[0],color='#EF5F3A')
    
    subplot(gs[1])
    plot(trime_sub['Moist(%)'].values,-trime_sub['RealDepth'].values, color='#EF5F3A',label='TDR')
    plot(30.*(1.+GPR_b[100:,int(idx/xres)]),-(np.arange(len(GPR_b[100:,1]))+100)*100.*caliparam, color='#47CEF7',label='GPR')
    plot(30.*(1.+GPR_g[100:,int(idx/xres)]),-(np.arange(len(GPR_g[100:,1]))+100)*100.*caliparam, color='#BEF363',label='GPRold')   
    sns.despine(offset=5, trim=False)
    plt.yticks([0, -100, -200, -300, -400],[0, -1, -2, -3, -4])
    plt.xticks([0, 10, 20, 30])
    text(latadj*30,100*(-0+vertadj),0,horizontalalignment='right',color='#366AC3')
    text(latadj*30,100*(-1+vertadj),int(np.round(1./(v_soil_eff*0.5))),horizontalalignment='right',color='#366AC3')
    text(latadj*30,100*(-2+vertadj),int(np.round(2./(v_soil_eff*0.5))),horizontalalignment='right',color='#366AC3')
    text(latadj*30,100*(-3+vertadj),int(np.round(3./(v_soil_eff*0.5))),horizontalalignment='right',color='#366AC3')
        
    sns.axlabel('theta [%]','')
    leg=plt.legend(loc='lower center')
    frame=leg.set_frame_on(True) 
    text(0, 10,'Soil Moisture Log')
```

```python
# get dtheta as time series at a TDR log:
def dtrime(TDRlog):
    subtrime=trime[(trime.MeasPoint=='t11')].copy()
    tgroup={}
    it=0
    zname=subtrime['RoundDepth'].unique()
    for name, group in subtrime.groupby('RoundDepth')['Moist(%)']:
        tgroup[zname[it]]=group-group[np.argmax(group.index)]
        it+=1
        
    df=pd.DataFrame(tgroup[zname[0]].resample('H', how='min').interpolate())
    for i in np.arange(len(zname))[1:]:
        df2=pd.DataFrame(tgroup[zname[i]].resample('H', how='min').interpolate())
        df=pd.concat([df,df2], join='inner', axis = 1)
    df.columns=zname
    return df
```

```python
def TS_GPR(i,TDRtube):
    import scipy.ndimage.filters as spf
    tube=['T'+str(TDRtube),'t'+str(TDRtube)]
    GPR_dummy=np.zeros(np.append(res, 8))
    GPR_bg=np.zeros(res)
    
    #get GPR profiles
    for k in np.arange(8):
        GPR_in=loadGPR('./GPR_data/line_data/'+files[k+(i*8)],res)
        GPR_dummy[:,:,k]=spf.median_filter(GPR_in,3)
    
    for k in np.arange(len(GPR_dummy[:,0,0])):
        for j in np.arange(len(GPR_dummy[0,:,0])):
            if len(np.where(diff(GPR_dummy[k,j,:])<=-0.05)[0])==0:
                GPR_bg[k,j]=9
            else:
                if i>0:
                    GPR_bg[k,j]=np.amax([2,np.where(diff(GPR_dummy[k,j,:])<=-0.05)[0][0]+2])
                else:
                    GPR_bg[k,j]=np.where(diff(GPR_dummy[k,j,:])<=-0.05)[0][0]
    
    #find closest stretch to TDR log
    ref=posi[posi.Point==tube[0]].values[0][[2,1]]
    dist=np.zeros(len(GPR1pos))
    for ix in np.arange(len(GPR1pos)):
        dist[ix]=np.linalg.norm(GPR1pos[ix,:2]-ref)
    idx=np.argmin(dist)
    
    GPR_d=np.zeros([8,3])
    GPR_d[0,0]=np.sum(1.-GPR_dummy[80:300,idx,0])
    GPR_d[0,1]=np.sum(1.-GPR_dummy[80:300,idx,0])
    
    for k in np.arange(8)[1:]:
        GPR_g = np.ma.array(1.-GPR_dummy[80:300,:,k], mask=GPR_bg[80:300,:]<=k)
        GPR_b = np.ma.array(1.-GPR_dummy[80:300,:,k], mask=GPR_bg[80:300,:]>k)
        GPR_d[k,0]=np.sum(1.-GPR_dummy[80:300,idx,k])
        GPR_d[k,1]=np.sum(GPR_g[:,idx])
        GPR_d[k,2]=np.sum(GPR_b[:,idx])
    
    GPR_d=pd.DataFrame(GPR_d)
    GPR_d.index=GPRtiming.index[GPRtiming.Profile==(i+1)]
    GPR_d.columns=['total','old','exp']
    
    dTDR=dtrime(tube[1])
    dTDR_TS=dTDR.sum(axis=1)
    
    jointDF=pd.concat([pd.DataFrame(dTDR_TS),GPR_d.resample('1h', how='mean').interpolate()], join='inner', axis = 1)
    jointDF.columns=['dTDR','dGPR_t','dGPR_o','dGPR_x']
    
    with sns.axes_style("white"):
        fig = plt.figure()
        ax1 = fig.add_subplot(111)
        ax1.plot(jointDF.dTDR,color='#EF5F3A',label='TDR')
        ax2 = ax1.twinx()
        ax2.plot(jointDF.dGPR_o,color='#47CEF7',label='GPR_o')
        ax2.plot(jointDF.dGPR_x,color='#BEF363',label='GPR_x')
        ax2.plot(jointDF.dGPR_t,color='#F7E917',label='GPR_t')
        ax2.plot(0, 0, '#EF5F3A',label='TDR')
        ax2.set_yticks(np.linspace(ax2.get_yticks()[0],ax2.get_yticks()[-1],len(ax1.get_yticks())))
        sns.despine(offset=5, trim=False, right=False)
        plt.legend(loc=1)
    
    return [jointDF,GPR_dummy,GPR_bg] #this is for debugging
    
```
