```python
## Catchements sizes in m2
Area=pd.DataFrame(np.array([19033276.859635,451211.742,1063268.993,1562816.821,214976.928,92475.551,459105.388,5462202.539,5254594.678,246722920.542,2700000.,4500000.]),
                  index=['Colpach','Weierbach','F','J','L','R','U','X','Z','Useldange','Huewelerbach','Wollefsbach']).T
```

```python
## standard data directory
datadir = '../Cluster Data/Holtzbach_current/'
```

```python
#read cluster U discharge data
CDT_U=pd.read_csv(datadir+'../S_U/S_U_CTDY.dat',sep='\t',na_values='NA')
CDT_U.index=pd.to_datetime(CDT_U['TIMESTAMP'],format='%Y-%m-%d %H:%M:%S')
CDT_U=CDT_U.drop(['TIMESTAMP'], 1)
# Apply rating curve to convert observed water level into discharge
h0=16.0
Q_U=(1.4*((CDT_U.H2OLevl_Y_mm_Avg-h0)/1000)**(5/2))*1000  #[l/s]
Q_U*=3.6/Area.U.values              #[m/hr]
Q_U*=1000.                           #[mm/hr]
```

```python
#read cluster R discharge data
CDT_R=pd.read_csv(datadir+'../S_R/S_R_CTD.dat',sep='\t',na_values='NA')
CDT_R.index=pd.to_datetime(CDT_R['TIMESTAMP'],format='%Y-%m-%d %H:%M:%S')
CDT_R=CDT_R.drop(['TIMESTAMP'], 1)
# Apply rating curve to convert observed water level into discharge
h0=30.5
Q_R=(0.775*((CDT_R.H2OLevl_X_mm_Avg-h0)/1000)**(2.47))*1000  #[l/s]
Q_R*=3.6/Area.R.values              #[m/hr]
Q_R*=1000.                           #[mm/hr]
```

```python
#read cluster J discharge data
CDT_J=pd.read_csv(datadir+'../S_J/S_J_CTD.dat',sep='\t',na_values='NA')
CDT_J.index=pd.to_datetime(CDT_J['TIMESTAMP'],format='%Y-%m-%d %H:%M:%S')
CDT_J=CDT_J.drop(['TIMESTAMP'], 1)
# Apply rating curve to convert observed water level into discharge
h0=165

Q_J=((2./3.*0.52*(np.sqrt(2.*9.81))*0.45*(((CDT_J.H2OLevl_X_mm_Avg-h0)/1000.)**1.5)*(1.+((4.*((CDT_J.H2OLevl_X_mm_Avg-h0)/1000.))/(5.*(0.45)/(0.839)))))*1000.)
Q_J[CDT_J.H2OLevl_X_mm_Avg>170.]=(3.953*(CDT_J.H2OLevl_X_mm_Avg[CDT_J.H2OLevl_X_mm_Avg>170.]/1000.)-0.3518)
Q_J*=3.6/Area.J.values              #[m/hr]
Q_J*=1000.                           #[mm/hr]
Q_J[Q_J<0.0005]=np.nan
```

```python
#somehow the rating curve appears not trustworthy. the gauge cannot be used here.
```

```python
#read cluster F discharge data
CDT_F=pd.read_csv(datadir+'../S_F/S_F_CTD.dat',sep='\t',na_values='NA')
CDT_F.index=pd.to_datetime(CDT_F['TIMESTAMP'],format='%Y-%m-%d %H:%M:%S')
CDT_F=CDT_F.drop(['TIMESTAMP'], 1)
# Apply rating curve to convert observed water level into discharge
h0=43
Q_F=0.04698076 *((CDT_F.H2OLevl_X_mm_Avg-h0)/10)**2.45415 #[l/s]
Q_F*=3.6/Area.F.values              #[m/hr]
Q_F*=1000.                           #[mm/hr]
```

```python
#read further Attert discharge data
def read_Q(filename):
    catchment=filename.split('/')[2].split('-')[1]
    dummy=pd.read_csv(filename,sep='\t',na_values='---',header=-1)
    dummy.index=pd.to_datetime(dummy[0]+' '+dummy[1],format='%d/%m/%Y %H:%M:%S')
    dummy.columns=['date','time',catchment]
    return dummy.iloc[:,2:]

Qfiles=['Q-Colpach-2001-2005-[m3.s-1].txt','Q-Colpach-2006-2010-[m3.s-1].txt','Q-Colpach-2011-2013-[m3.s-1].txt',
        'Q-Huewelerbach-2001-2010-[m3.s-1].txt','Q-Useldange-1998-2010-[m3.s-1].txt','Q-Weierbach-2004-2010-[m3.s-1].txt',
        'Q-Weierbach-2011-2013-[m3.s-1].txt','Q-Wollefsbach-2001-2010-[m3.s-1].txt']

Qdir='../obs_discharge-Attert-basin_CRPGL/'

#get list of all covered catchments
dummyc=[]
for i in np.arange(len(Qfiles)):
    dummyc=np.append(dummyc,Qfiles[i].split('-')[1])

#COLPACH
idx=np.where(dummyc=='Colpach')[0]
Q_Colpach=read_Q(Qdir+Qfiles[idx[0]])
for i in np.arange(len(idx))[1:]:
    Q_Colpach=pd.concat([Q_Colpach,read_Q(Qdir+Qfiles[idx[i]])])

#HUEWELERBACH
idx=np.where(dummyc=='Huewelerbach')[0]
Q_Huewelerbach=read_Q(Qdir+Qfiles[idx[0]])
for i in np.arange(len(idx))[1:]:
    Q_Huewelerbach=pd.concat([Q_Huewelerbach,read_Q(Qdir+Qfiles[idx[i]])])
#apparently the rating curve has not been applied to the data yet, used 2011 RC!
Q_Huewelerbach=1.7784*((Q_Huewelerbach-8)/100)**1.3123
        
#WOLLEFSBACH
idx=np.where(dummyc=='Wollefsbach')[0]
Q_Wollefsbach=read_Q(Qdir+Qfiles[idx[0]])
for i in np.arange(len(idx))[1:]:
    Q_Wollefsbach=pd.concat([Q_Wollefsbach,read_Q(Qdir+Qfiles[idx[i]])])

#WEIERBACH
idx=np.where(dummyc=='Weierbach')[0]
Q_Weierbach=read_Q(Qdir+Qfiles[idx[0]])
for i in np.arange(len(idx))[1:]:
    Q_Weierbach=pd.concat([Q_Weierbach,read_Q(Qdir+Qfiles[idx[i]])])

#USELDANGE
idx=np.where(dummyc=='Useldange')[0]
Q_Useldange=read_Q(Qdir+Qfiles[idx[0]])
for i in np.arange(len(idx))[1:]:
    Q_Useldange=pd.concat([Q_Useldange,read_Q(Qdir+Qfiles[idx[i]])])
```

```python
#get additional data for Colpach and Weierbach
dummy=pd.read_csv('../Cluster Data/CAOS-Q-Colpach-2010-2013.txt',sep=';',header=-1,skiprows=3)
dummy.index=pd.to_datetime(dummy[0]+' '+dummy[1],format='%d/%m/%Y %H:%M:%S')
dummy.columns=['date','time','Colpach']
Q_Colpach2=dummy.iloc[:,2:]

dummy=pd.read_csv('../Cluster Data/CAOS-Q-Weierbach-2010-2013.txt',sep=';',header=-1,skiprows=3,na_values='---')
dummy.index=pd.to_datetime(dummy[0]+' '+dummy[1],format='%d/%m/%Y %H:%M:%S')
dummy.columns=['date','time','Weierbach']
Q_Weierbach2=dummy.iloc[:,2:]
```
