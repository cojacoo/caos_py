```python
#load profile file
def loadGPR(filex,res):
    GPR_dummy=np.loadtxt(filex)
    GPR_z=np.ones(res)
    wx=np.amin([shape(GPR_dummy)[1],shape(GPR_z)[1]])
    GPR_z[:,:wx]=GPR_dummy[:res[0],:wx]
    return GPR_z
```

# Analyse dynamics of each pixel
For each pixel in a sequence of profiles its first decline (which corresponds to increased wetting) is determined. As threshold a change of the difference attribute of 0.05 is taken. Declining structures are marked green. Inclining ones in blue. We assume that green corresponds to active structures from the natural storm event and blue to structures which have been activated by our sprinkling experiment.

```python
import scipy.ndimage.filters as spf

def TS_GPR(i):
    GPR_dummy=np.zeros(np.append(res, 8))
    GPR_bg=np.zeros(res)
    for k in np.arange(8):
        GPR_in=loadGPR('./GPR_data/line_data/'+files[k+(i*8)],res)
        GPR_dummy[:,:,k]=spf.median_filter(GPR_in,3)
    
    for k in np.arange(len(GPR_dummy[:,0,0])):
        for j in np.arange(len(GPR_dummy[0,:,0])):
            if len(np.where(diff(GPR_dummy[k,j,:])<=-0.05)[0])==0:
                GPR_bg[k,j]=9
            else:
                if i>0:
                    GPR_bg[k,j]=np.amax([2,np.where(diff(GPR_dummy[k,j,:])<=-0.05)[0][0]+2])
                else:
                    GPR_bg[k,j]=np.where(diff(GPR_dummy[k,j,:])<=-0.05)[0][0]
    
    subplot(gs[i])
    plt.imshow(-GPR_dummy[:,:,0],cmap='Greens',aspect=1,extent=[0,10,-4,0],vmin=-1., vmax=0.)
    sns.despine(offset=5, trim=False)
    plt.yticks([0, -1, -2, -3, -4])
    text(7.5, 0.1,'time: '+files[0+(i*8)].split()[3].split('h')[0])
    text(0.1, 0.1,'Profile '+str(i+1))
    if i==0:
        sns.axlabel('','depth [m]')
    
    for k in np.arange(8)[1:]:
        subplot(gs[i+k*4])
        GPR_g = np.ma.array(-GPR_dummy[:,:,k], mask=GPR_bg<=k)
        GPR_b = np.ma.array(-GPR_dummy[:,:,k], mask=GPR_bg>k)
        plt.imshow(GPR_g,cmap='Greens',aspect=1,extent=[0,10,-4,0],vmin=-1., vmax=0.,interpolation='none')
        plt.imshow(GPR_b,cmap='Blues',aspect=1,extent=[0,10,-4,0],vmin=-1., vmax=0.,interpolation='none')
        sns.despine(offset=5, trim=False)
        plt.yticks([0, -1, -2, -3, -4])
        text(7.5, 0.1,'time: '+files[k+(i*8)].split()[3].split('h')[0])
        text(0.1, 0.1,'Profile '+str(i+1))
        if i==0:
            sns.axlabel('','depth [m]')
        if k==7:
            sns.axlabel('lateral position on profile line [m]','')
        if (i==0) & (k==7):
            sns.axlabel('lateral position on profile line [m]','depth [m]')
```
