```python
## standard data directory
datadir = '../Cluster Data/Holtzbach_current/'
```

```python
#read cluster U discharge data
CDT_U=pd.read_csv('/Users/cojack/Documents/phd/caos/DATA/Colpach/S_U/S_U_CTDY.dat',sep='\t',na_values='NA')
CDT_U.index=pd.to_datetime(CDT_U['TIMESTAMP'],format='%Y-%m-%d %H:%M:%S')
CDT_U=CDT_U.drop(['TIMESTAMP'], 1)
```

```python
# Apply rating curve to convert observed water level into discharge
h0=16.0
Q_U=(1.4*((CDT_U.H2OLevl_Y_mm_Avg-h0)/1000)**(5/2))*1000  #[l/s]
Q_U*=3.6/Areas.U.values              #[m/hr]
Q_U*=1000.                           #[mm/hr]
```

```python
#read cluster R discharge data
CDT_R=pd.read_csv(datadir+'S_R_CTD.dat',sep='\t',na_values='NA')
CDT_R.index=pd.to_datetime(CDT_R['TIMESTAMP'],format='%Y-%m-%d %H:%M:%S')
CDT_R=CDT_R.drop(['TIMESTAMP'], 1)
# Apply rating curve to convert observed water level into discharge
h0=30.5
Q_R=(0.775*((CDT_R.H2OLevl_X_mm_Avg-h0)/1000)**(2.47))*1000  #[l/s]
Q_R*=3.6/Areas.R.values              #[m/hr]
Q_R*=1000.                           #[mm/hr]
```

```python
# concat both time series:
Q_holtz=pd.concat([Q_U,Q_R], join='inner', axis = 1)
Q_holtz.columns=['Q_U','Q_R']
Q_holtz.resample('1h', how='mean').plot()
```

```python
# Reading the 5 minutes timestep rainfall times series of the cluster I 
Precip_I=pd.read_csv(datadir+'S_I_meteo/S_I_rain.dat',sep='\t',na_values='NA')
Precip_I.index=pd.to_datetime(Precip_I['TIMESTAMP'],format='%Y-%m-%d %H:%M:%S')
Precip_I=Precip_I.drop(['TIMESTAMP'], 1)
# Aggregating the 5 minutes timestep rainfall into hourly timestep.
Precip_I.resample('1h', how='sum').plot()
```

```python
# Reading the 5 minutes timestep soil moisture times series of the cluster V 
Theta_V=pd.read_csv(datadir+'S_V_FiveTE.dat',sep='\t',na_values='NA')
Theta_V.index=pd.to_datetime(Theta_V['TIMESTAMP'],format='%Y-%m-%d %H:%M:%S')
Theta_V=Theta_V.drop(['TIMESTAMP'], 1)
Theta_V=Theta_V.drop(Theta_V.columns[9:], 1)
Theta_V.columns=['SM10a','SM30a','SM50a','SM10b','SM30b','SM50b','SM10c','SM30c','SM50c']
# clean errors
Theta_V[Theta_V>0.8]=np.nan
Theta_V[Theta_V<0.08]=np.nan
Theta_V.resample('1h', how='mean').plot()
```

```python
# concat time series:
WB_holtz=pd.concat([Q_holtz,Theta_V,Precip_I], join='inner', axis = 1)
WB_holtz.rename(columns={'rain_mm_Tot.1.':'Prec'}, inplace=True)

mlab=['Q_U','Q_R','$\theta$ 0.1 m','$\theta$ 0.3 m','$\theta$ 0.5 m']
mlabcols=['Q_U','Q_R','SM10a','SM30a','SM50a']
```
